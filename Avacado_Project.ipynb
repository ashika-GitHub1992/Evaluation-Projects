# Removing Outliers:

z-Score Technique:

```python
from scipy.stats import zscore
import numpy as np
z=np.abs(zscore(avc))
z.shape
```

```python
threshold=3
print(np.where(z>3))
```

```python
loss_percent=(18249-17714)/18249*100
print(loss_percent)
```

IQR Technique:

```python
q1=avc.quantile(0.25)
q3=avc.quantile(0.75)

IQR=q3-q1
print(IQR)
```

```python
avc_new1=avc[~((avc<(q1-1.5*IQR)) | (avc>(q3+1.5*IQR))).any(axis=1)]
print(avc_new1.shape)
```

Percentage Data Loss:

```python
loss_percent=(18249-12128)/18249*100
print(loss_percent)
```

The percentage loss in the IQR is more. So we prefer z-score over IQR.

```python
for i in range(0,1000):
    x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=i,test_size=0.20)
    lr.fit(x_train,y_train)
    pred_train=lr.predict(x_train)
    pred_test=lr.predict(x_test)
    if round(r2_score(y_train,pred_train)*100,1)==round(r2_score(y_test,pred_test)*100,1):
        print("At random state:",i)
        print("Training r2_score is:",r2_score(y_train,pred_train*100))
        print("Testing r2_score is:",r2_score(y_test,pred_test*100))
```
