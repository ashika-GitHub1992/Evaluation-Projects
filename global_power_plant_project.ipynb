# Label Encoding:

```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()

gppd['country']=le.fit_transform(gppd['country'])
gppd['country_long']=le.fit_transform(gppd['country_long'])
gppd['name']=le.fit_transform(gppd['name'])
gppd['gppd_idnr']=le.fit_transform(gppd['gppd_idnr'])
gppd['primary_fuel']=le.fit_transform(gppd['primary_fuel'])
gppd['other_fuel1']=le.fit_transform(gppd['other_fuel1'])
gppd['other_fuel2']=le.fit_transform(gppd['other_fuel2'])
gppd['other_fuel3']=le.fit_transform(gppd['other_fuel3'])
gppd['owner']=le.fit_transform(gppd['owner'])
gppd['source']=le.fit_transform(gppd['source'])
gppd['url']=le.fit_transform(gppd['url'])
gppd['geolocation_source']=le.fit_transform(gppd['geolocation_source'])
gppd['wepp_id']=le.fit_transform(gppd['wepp_id'])
gppd['generation_data_source']=le.fit_transform(gppd['generation_data_source'])
gppd['estimated_generation_gwh']=le.fit_transform(gppd['estimated_generation_gwh'])

gppd
```

Filling NaN values:

```python
import numpy as np
gppd=gppd.replace(np.NaN,gppd['latitude'].mean())
gppd=gppd.replace(np.NaN,gppd['longitude'].mean())
gppd=gppd.replace(np.NaN,gppd['other_fuel1'].mode())
gppd=gppd.replace(np.NaN,gppd['other_fuel2'].mode())
gppd=gppd.replace(np.NaN,gppd['other_fuel3'].mode())
gppd=gppd.replace(np.NaN,gppd['commissioning_year'].mean())
gppd=gppd.replace(np.NaN,gppd['owner'].mode())
gppd=gppd.replace(np.NaN,gppd['geolocation_source'].mode())
gppd=gppd.replace(np.NaN,gppd['wepp_id'].mode())
gppd=gppd.replace(np.NaN,gppd['year_of_capacity_data'].mean())
gppd=gppd.replace(np.NaN,gppd['generation_gwh_2013'].mean())
gppd=gppd.replace(np.NaN,gppd['generation_gwh_2014'].mean())
gppd=gppd.replace(np.NaN,gppd['generation_gwh_2015'].mean())
gppd=gppd.replace(np.NaN,gppd['generation_gwh_2016'].mean())
gppd=gppd.replace(np.NaN,gppd['generation_gwh_2017'].mean())
gppd=gppd.replace(np.NaN,gppd['generation_data_source'].mode())
gppd=gppd.replace(np.NaN,gppd['estimated_generation_gwh'].mode())
```

# Removing Outliers:

z-Score Technique:

```python
from scipy.stats import zscore
import numpy as np
z=np.abs(zscore(gppd))
z.shape
```

```python
threshold=3
print(np.where(z>3))
```

```python
gppd_new=gppd[(z<3).all(axis=1)]
print(gppd.shape)
print(gppd_new.shape)
```

Percentage Data Loss:

```python
loss_percent=(908-821)/908*100
print(loss_percent)
```

IQR Technique:

```python
q1=gppd.quantile(0.25)
q3=gppd.quantile(0.75)

IQR=q3-q1
print(IQR)
```

```python
gppd_new1=gppd[~((gppd<(q1-1.5*IQR)) | (gppd>(q3+1.5*IQR))).any(axis=1)]
print(gppd.shape)
print(gppd_new1.shape)
```

# Linear Regression:

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
from sklearn.metrics import r2_score
```

```python
for i in range(0,1000):
    x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=i,test_size=0.20)
    lr.fit(x_train,y_train)
    pred_train=lr.predict(x_train)
    pred_test=lr.predict(x_test)
    if round(r2_score(y_train,pred_train)*100,1)==round(r2_score(y_test,pred_test)*100,1):
        print("At random state:",i)
        print("Training r2_score is:",r2_score(y_train,pred_train*100))
        print("Testing r2_score is:",r2_score(y_test,pred_test*100))
```
